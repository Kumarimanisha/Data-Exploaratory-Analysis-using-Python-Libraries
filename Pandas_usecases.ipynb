## READING JSON FILE AS PANDAS DATAFRAME

```python
json_df=pd.read_json('dwsample1-json.json',orient='records')
add_in_json=pd.DataFrame({"Courses":"English","Fee":"15000","Duration":"90 Days","Discount":"1800"},index=[4])
json_df=pd.concat([json_df,add_in_json])
```

```python
json_df
```

```python
json_df.drop(json_df.index[json_df['Courses']=='English'],inplace=True) #deleting row based on condition
```

```python
json_df.info()
json_df['Fee']=json_df['Fee'].astype(int)
```

```python
json_df.nlargest(3,'Fee',keep='first')#keeps only frist record in case of tie. use keep='all' to keep all the records of partition
```

# PIVOTING DATAFRAME

```python
j_pivoted=json_df.pivot(index='Courses',columns='Duration',values=['Fee','Discount'])
j_pivoted
```

```python
j_pivoted_table=json_df.pivot_table(index='Courses',columns='Duration',values=['Fee','Discount'])
j_pivoted_table
```
